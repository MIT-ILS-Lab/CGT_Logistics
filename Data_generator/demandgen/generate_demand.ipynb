```python
# Parameters
n_patients = 1800
num_rows = int(n_patients/4)

# Generate random data for 'c' (2nd column) between c1 and c4
categories = ["c1", "c2", "c3", "c4"]
random_c = np.random.choice(categories, size=num_rows)

# Generate random data for 'value1' (3rd column) with values between 1 and 90
random_value1 = np.random.randint(1, 91, size=num_rows)  # Uniformly distributed between 1 and 90

# Create the DataFrame
updated_data = {
    "p": [f"p{i}" for i in range(1, num_rows + 1)],
    "c": random_c,
    "value1": random_value1,
    "value2": [1] * num_rows
}

updated_df = pd.DataFrame(updated_data)

# Display the DataFrame
print(updated_df)

```
